```python
import subprocess
import os
import shutil
```

## Pulling data from SCIL server

```python
url = "https://scil-data.sdsc.edu/data/ul-risk-modeling-2026/FuelMaps/Tahoe/"
username = ""
password = ""
output_dir = "."

os.makedirs(output_dir, exist_ok=True)

subprocess.run([
    "wget",
    f"--user={username}",
    f"--password={password}",
    "--recursive",
    "--no-parent",
    "--no-host-directories",
    "--cut-dirs=3",
    '--reject=index.html*',
    url
], cwd=output_dir, check=True)
```

```python
# Rename Tahoe to Inputs - This is needed for the trees.exe
if os.path.exists("Tahoe") and not os.path.exists("Inputs"):
    os.rename("Tahoe", "Inputs")
```
