## Pulling data from SCIL server

```python
import subprocess
import os

url = "https://scil-data.sdsc.edu/data/ul-risk-modeling-2026/FuelMaps/Tahoe/"
username = "amarillo"
password = "tahoe"
output_dir = "."

os.makedirs(output_dir, exist_ok=True)

subprocess.run([
    "wget",
    f"--user={username}",
    f"--password={password}",
    "--recursive",
    "--no-parent",
    "--no-host-directories",
    "--cut-dirs=3",
    '--reject=index.html*',
    url
], cwd=output_dir, check=True)
```
